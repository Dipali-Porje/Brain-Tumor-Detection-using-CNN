```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
```

```python
train_dir = r"C:\Users\porje\OneDrive\Desktop\internship jb infotech\CNN\brain tumor prediction\Training"
test_dir = r"C:\Users\porje\OneDrive\Desktop\internship jb infotech\CNN\brain tumor prediction\Testing"
```

```python
img_size = (224, 224)  # Adjust based on your model requirement
batch_size = 32

train_datagen = ImageDataGenerator(
    rescale=1./255,  # Normalize pixel values
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

test_datagen = ImageDataGenerator(rescale=1./255)  # Only rescaling for testing

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical'  # Change to 'binary' for two classes
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical'
)
```

```python
model = keras.Sequential([
    # First block
    keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(224, 224, 3)),
    keras.layers.MaxPooling2D(2,2),

    # Second block
    keras.layers.Conv2D(64, (3,3), activation='relu'),
    keras.layers.MaxPooling2D(2,2),

    # Third block (newly added)
    keras.layers.Conv2D(128, (3,3), activation='relu'),
    keras.layers.MaxPooling2D(2,2),

    # Fourth block (newly added)
    keras.layers.Conv2D(256, (3,3), activation='relu'),
    keras.layers.MaxPooling2D(2,2),

    keras.layers.Flatten(),
    keras.layers.Dense(256, activation='relu'),
    keras.layers.Dropout(0.5),  # Reduce overfitting
    keras.layers.Dense(train_generator.num_classes, activation='softmax')
])
```

```python
from tensorflow.keras.models import load_model

model = load_model(r'C:\Users\porje\OneDrive\Desktop\internship jb infotech\CNN\brain tumor prediction\brain_tumor_model2.h5')
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(
    train_generator,
    validation_data=test_generator,
    epochs=10
)
```

```python
model.save('brain_tumor_model2.h5')
```
